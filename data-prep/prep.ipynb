### Create embeddings for quering the similarity between the jokes more optimally

```python
import pandas as pd
from sentence_transformers import SentenceTransformer

df = pd.read_csv('jokes.csv', sep=',')

# Create embeddings
model = SentenceTransformer('all-mpnet-base-v2')
jokes = df['jokeText'].tolist()
embeddings = model.encode(jokes, convert_to_numpy=True)

print(embeddings)
```
